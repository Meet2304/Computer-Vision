```python
from PIL import Image
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
Img = Image.open("Images/Rocket_Image.jpg")
img1 = cv2.imread("Images/Rocket_Image.jpg")
```

```python
# Figure to compare default color mode in Opencv and PIL
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(Img)
plt.title("PIL Image")
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(img1)
plt.title("OpenCV default Image")
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title("OpenCV Image with RGB conversion")
plt.axis('off')
plt.show()
```

```python
# To display the file format of the image using PIL
print(f"Image file format: {Img.format}")
```

```python
# To display the pixel format of the image
print(f"Image pixel format: {Img.mode}")
```

```python
# To display the size of the image
print(f"Image size: {Img.size}")
```

```python
# Convert image from RGB to Grayscale using PIL
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
Gray_Img = Img.convert("L")
plt.imshow(Gray_Img, cmap='gray')
plt.title("Grayscale Image using PIL")
plt.axis('off')
# Convert image from RGB to Binary using PIL
Binary_Img = Img.convert("1")
plt.subplot(1, 2, 2)
plt.imshow(Binary_Img, cmap='gray')
plt.title("Binary Image using PIL")
plt.axis('off')
plt.show()

```

```python
# Resizing the image using PIL
resized_img = Img.resize((300, 300))
plt.imshow(resized_img) 
plt.title("Resized Image using PIL")
plt.axis('off')
plt.show()
```

```python
# Cropping the image using PIL
cropped_img = Img.crop((50, 50, 250, 250))
plt.figure(figsize=(5, 5))
plt.title("Cropped Image using PIL")
plt.axis('off')
plt.imshow(cropped_img)
```

```python
# Thumbnailing the image using PIL
thumbnail_img = Img.copy()
thumbnail_img.thumbnail((100, 100))
plt.figure(figsize=(5, 5))
plt.title("Thumbnail Image using PIL")
plt.axis('off')
plt.imshow(thumbnail_img)
plt.show()
```

```python
original_image = Image.open("Images/Rocket_Image.jpg")
# Splitting the imae into RGB channels
r, g, b = original_image.split()
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(r, cmap='Reds')
plt.title("Red Channel")
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(g, cmap='Greens')
plt.title("Green Channel")
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(b, cmap='Blues')
plt.title("Blue Channel")
plt.axis('off')
plt.show()
```

```python
# Merging the RGB channels back into a single image
merged_image = Image.merge("RGB", (r, g, b))
plt.imshow(merged_image)
plt.title("Merged RGB Channels")
plt.axis('off')
plt.show()
```

```python
# Saving the merged image
merged_image.save("Images/Merged_RGB_Image.jpg")
```
